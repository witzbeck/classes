```python
# for dist
df = 1
rv = chi2(df)
mean, var, skew, kurt = chi2.stats(df, moments='mvsk')
x_n = 100
x_min = chi2.ppf(0.01, df)
x_max = chi2.ppf(0.99, df)
x = linspace(0, 1, x_n)
#x = linspace(x_min, x_max, x_n)
#y = chi2.pdf(x, df)
y = rv.pdf(x)
def chi2_dist(dof, x_n: int = 10000):
    rand_var = chi2(dof)
    rand_vals = chi2.rvs(dof, size=x_n)
    x_min = chi2.ppf(0.01, dof)
    x_max = chi2.ppf(0.99, dof)
    x = linspace(x_min, x_max, x_n)
    y = rand_var.pdf(x)
    #fig, ax = subplots(1,1)
    #ax.set_xlim(x_min, x_max)
    #ax.hist(rand_vals, density=True, bins="auto", histtype="stepfilled", alpha=0.2)
    #ax.plot(x, y)
    return x, y, rand_vals, x_min, x_max
chi2_dist(100)
```

```python
fig, ax = subplots(1,1)

dof0, dof1, dof2, dof3 = 10, 100, 1000, 10000

x, y, r, x_min, x_max = chi2_dist(dof3)
ax.set_xlim(x_min, x_max)
ax.hist(r, density=True, bins="auto", histtype="stepfilled", alpha=0.2)
ax.plot(x, y)
```
