```python
def convert_numeric(bits):
    bits = bits.strip("0")
    bits0 = [x.count("0") for x in bits.split("1") if x != ""]
    bits1 = [x.count("1") for x in bits.split("0") if x != ""]
    brange = range(len(bits0))
    new_list = []
    for i in brange:
        new_list.append(bits1[i])
        new_list.append(bits0[i])
    new_list.append(bits1[-1])
    return new_list


def simp(nums):
    primes = [2, 3, 5, 7, 11]
    for prime in primes:
        if 1 in nums:
            return nums
        elif (2 in nums and 3 in nums):
            return nums
        elif min([x % prime == 0 for x in nums])==1:
            nums = [x // prime for x in nums]
    return nums
def simp2(nums):
    _min = min(nums)
    return [x // _min for x in nums]

def decodeBitsAdvanced(bits_str):
    nums = convert_numeric(bits_str)
    simp_nums = simp2(nums)
    ret_str = ""
    isone = True
    for y in simp_nums:
        if (isone and y in (1, 2, 3)):
            ret_str = ret_str + "."
        elif (isone and y >= 3):
            ret_str = ret_str + "-"
        elif (isone is False and y in (1, 2)):
            pass
        elif (isone is False and y in (3, 4, 5)):
            ret_str = ret_str + " "
        elif (isone is False and y > 5):
            ret_str = ret_str + "   "
        elif isone is True:
            raise ValueError(isone, y, simp_nums)
        elif isone is False:
            raise ValueError(isone, y, simp_nums)
        else:
            raise ValueError(isone, y, simp_nums)

        if isone:
            isone = False
        else:
            isone = True
    return ret_str


def decodeMorse(morse_code):
    mc_split = morse_code.strip(" ").split(" ")
    return "".join([mc[x] for i, x in enumerate(mc_split) if not (mc_split[i-1] == "" and x == "")])
decode_bits("0000000011011010011100000110000001111110100111110011111100000000000111011111111011111011111000000101100011111100000111110011101100000100000")
```
